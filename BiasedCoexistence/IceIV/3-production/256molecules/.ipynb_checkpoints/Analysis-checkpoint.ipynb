```python
font = {'family' : 'serif',
        'serif'   : 'palatino',
        #'sans-serif'    : 'Computer Modern Sans serif',
        'style'   : 'normal',
        'variant'   : 'normal',
        'stretch'   : 'normal',
        'weight'   : 'normal',
        'size'   : 9}
plt.rc('font', **font)
plt.rc('text', usetex=True)
plt.rcParams['figure.figsize'] = (7, 0.5)

pressures=np.array([2500,3000,3500,4000])
temps=np.array([260,265,270,275,280,285,290])

def func(x,a,b):
    return a*x+b

fig, axs = plt.subplots(1,pressures.shape[0],sharex=True,sharey=True)

counter1=0
for press in pressures:
    counter2=0
    for temp in temps:
        file='IceIV-' + str(temp) + ".0-" + str(press) + '.0/COLVAR.0'
        if(os.path.exists(file)):
            data = np.genfromtxt(file)
            ignore=int(data.shape[0]/4)
            bias= data[ignore:,3]
            cv=data[ignore:,1]
            if (cv.shape[0]>0):
                beta=1./(0.00831441001626*temp)
                logweights=beta*bias
                logweights -= np.amax(logweights)
                nbins=20
                histmin=128
                histmax=192
                histo, bin_edges = np.histogram(cv,weights=np.exp(logweights),bins=nbins,range=(histmin,histmax))
                err = np.sqrt(np.histogram(cv,weights=np.power(np.exp(logweights),2),bins=nbins,range=(histmin,histmax))[0])
                bin_centers = (bin_edges[1:]+bin_edges[:-1])/2
                fes = -np.log(histo)
                offset = np.mean(np.ma.masked_invalid(fes))
                fes -= offset
                axs[counter1].set_xlim([histmin,histmax])
                axs[counter1].set_xticks([128,160,192])
                axs[counter1].set_ylim([-5,5])
                feserr = err/histo
                axs[counter1].fill_between(bin_centers,fes-10*feserr,fes+10*feserr,color=colors[counter2],linewidth=0,alpha=0.5)
            if (fes[np.isfinite(fes)].shape[0]>0):
                popt, pcov = curve_fit(func, bin_centers[np.isfinite(fes)], fes[np.isfinite(fes)], sigma=2*feserr[np.isfinite(fes)])
                x=np.linspace(histmin,histmax,10)
                axs[counter1].plot(x,func(x,*popt),'--',alpha=0.5,color=colors[counter2],linewidth=0.5)
        counter2 += 1
    axs[counter1].plot([histmin,histmax],[0,0],':',alpha=0.5,color='black',zorder=0,linewidth=0.5)
    counter1 += 1
    
for i in range(pressures.shape[0]):
    axs[i].text(160,6.6,str(pressures[i]/10000)+ ' GPa',va='center',ha='center')
    
axs[int(pressures.shape[0]/2)].xaxis.set_label_coords(-0.075, -0.6)
axs[int(pressures.shape[0]/2)].set_xlabel(r"\# of ice IV-like molecules")
axs[0].set_ylabel("F (k$_B$T)")

plt.savefig("2-fes-IV-small.pdf",bbox_inches='tight')  # publication quality

plt.show()
```

```python
font = {'family' : 'serif',
        'serif'   : 'palatino',
        #'sans-serif'    : 'Computer Modern Sans serif',
        'style'   : 'normal',
        'variant'   : 'normal',
        'stretch'   : 'normal',
        'weight'   : 'normal',
        'size'   : 9}
plt.rc('font', **font)
plt.rc('text', usetex=True)
plt.rcParams['figure.figsize'] = (7, 0.5)

colors = plt.rcParams['axes.prop_cycle'].by_key()['color']

#ignore=1*4000

def func2(x,a,x0):
    return a*(x-x0)

pressures=np.array([2500,3000,3500,4000])
temps=np.array([260,265,270,275,280,285,290])

fig, axs = plt.subplots(1,pressures.shape[0],sharex=True,sharey=True)
counter1=0
melt_temp=np.zeros(0)
melt_temp_errors=np.zeros(0)
melt_temp_pressures=np.zeros(0)
for press in pressures:
    counter2=0
    chem_pot=np.zeros(0)
    chem_pot_temps=np.zeros(0)
    for temp in temps:
        file='IceIV-' + str(temp) + ".0-" + str(press) + '.0/COLVAR.0'
        if(os.path.exists(file)):
            data = np.genfromtxt(file)
            ignore=int(data.shape[0]/4)
            bias=data[ignore:,3]
            cv=data[ignore:,1]
            if (cv.shape[0]>0):
                beta=1./(0.00831441001626*temp)
                logweights=beta*bias
                logweights -= np.amax(logweights)
                nbins=20
                histmin=128
                histmax=192
                histo, bin_edges = np.histogram(cv,weights=np.exp(logweights),bins=nbins,range=(histmin,histmax))
                err = np.sqrt(np.histogram(cv,weights=np.power(np.exp(logweights),2),bins=nbins,range=(histmin,histmax))[0])
                bin_centers = (bin_edges[1:]+bin_edges[:-1])/2
                fes = -np.log(histo)
                offset = np.mean(np.ma.masked_invalid(fes))
                fes -= offset
                feserr = err/histo
                popt, pcov = curve_fit(func, bin_centers[np.isfinite(fes)], fes[np.isfinite(fes)], sigma=2*feserr[np.isfinite(fes)])
                axs[counter1].scatter(temp,popt[0],s=10,color=colors[0])
                chem_pot=np.insert(chem_pot,0,popt[0])
                chem_pot_temps=np.insert(chem_pot_temps,0,temp)
        counter2 += 1
        axs[counter1].plot([255,295],[0,0],':',color='black',alpha=0.25,zorder=0,linewidth=0.5)
        axs[counter1].set_xlim([255,295])
        axs[counter1].set_xticks([260,290])
        axs[counter1].set_ylim([-0.25,0.25])
    if (chem_pot_temps.shape[0]>0):
        popt, pcov = curve_fit(func2, chem_pot_temps, chem_pot, sigma=np.ones(chem_pot.shape[0])*0.05, absolute_sigma=True)
        perr = np.sqrt(np.diag(pcov))
        x=np.linspace(temps[0]-10,temps[-1]+10,10)
        axs[counter1].plot(x,func2(x,*popt),'--',color='black',alpha=0.5,linewidth=0.75)
        melt_temp=np.insert(melt_temp,0,popt[1])
        melt_temp_pressures=np.insert(melt_temp_pressures,0,press)
        melt_temp_errors=np.insert(melt_temp_errors,0,perr[1])
        axs[counter1].fill_between(x,func2(x,popt[0],popt[1]-perr[1]),func2(x,popt[0],popt[1]+perr[1]),linewidth=0.,color='black',alpha=0.2,zorder=0)
        counter1 += 1
    
print(melt_temp,melt_temp_pressures,melt_temp_errors)


for i in range(pressures.shape[0]):
    axs[i].text(275,0.35,str(pressures[i]/10000)+ ' GPa',va='center',ha='center')
    

axs[int(pressures.shape[0]/2)].xaxis.set_label_coords(-0.075, -0.6)
axs[int(pressures.shape[0]/2)].set_xlabel("Temperature (K)")
axs[0].set_ylabel("$\Delta\mu$ (k$_B$T)")

plt.savefig("3-deltamu-IV-small.pdf",bbox_inches='tight')  # publication quality

plt.show()
```

```python
font = {'family' : 'serif',
        'serif'   : 'palatino',
        #'sans-serif'    : 'Computer Modern Sans serif',
        'style'   : 'normal',
        'variant'   : 'normal',
        'stretch'   : 'normal',
        'weight'   : 'normal',
        'size'   : 20}
plt.rc('font', **font)
#plt.rc('text', usetex=False)
plt.rcParams['figure.figsize'] = (15, 4)

colors = plt.rcParams['axes.prop_cycle'].by_key()['color']

pressures=np.array([1500,1750,2000,2500,3000,3500,4000])
temps=np.array([240,245,250,255,260,265,270,275,280,285,290])

def func2(x,a,x0):
    return a*(x-x0)

#ignore=1*4000
#pressures=np.array([2000,2500,3000,3500,4000])
#temps=np.array([255,260,265,270,275,280,285])
fig, axs = plt.subplots(1,pressures.shape[0],sharex=True,sharey=True)
counter1=0
melt_temp=np.zeros(0)
melt_entropy=np.zeros(0)
melt_temp_pressures=np.zeros(0)
melt_temp_errors=np.zeros(0)
for press in pressures:
    #filenamesList = sorted(glob.glob('IceIV*' +str(int(press)) + '*/COLVAR.0'))
    #print(filenamesList)
    counter2=0
    chem_pot=np.zeros(0)
    chem_pot_temps=np.zeros(0)
    for temp in temps:
        file='IceIV-' + str(temp) + ".0-" + str(press) + '.0/COLVAR.0'
        if(os.path.exists(file)):
            data = np.genfromtxt(file)
            ignore=int(data.shape[0]/2)
            bias=data[ignore:,3]
            cv=data[ignore:,1]
            if (cv.shape[0]>0):
                beta=1./(0.00831441001626*temp)
                logweights=beta*bias
                logweights -= np.amax(logweights)
                nbins=20
                histmin=128
                histmax=192
                histo, bin_edges = np.histogram(cv,weights=np.exp(logweights),bins=nbins,range=(histmin,histmax))
                err = np.sqrt(np.histogram(cv,weights=np.power(np.exp(logweights),2),bins=nbins,range=(histmin,histmax))[0])
                bin_centers = (bin_edges[1:]+bin_edges[:-1])/2
                fes = -(1/beta)*np.log(histo)
                offset = np.mean(np.ma.masked_invalid(fes))
                #print(offset)
                fes -= offset
                #axs[counter1,counter2].plot(bin_centers,fes,color=colors[counter1])
                #axs[counter1,counter2].set_xlim([128,192])
                #axs[counter1,counter2].set_xticks([128,192])
                #axs[counter1,counter2].set_ylim([-10,10])
                # from https://ddavis.io/posts/numpy-histograms/
                #b = np.linspace(histmin,histmax, nbins+1)
                #digits = np.digitize(cv, b)                 ## bin index array for each data element
                #sum_w2 = np.zeros([nbins], dtype=np.float32)
                #for i in range(nbins):
                #    weights_in_current_bin = np.exp(logweights)[np.where(digits == i)[0]]
                #    sum_w2[i] = np.sum(np.power(weights_in_current_bin, 2))
                #err = np.sqrt(sum_w2)
                #fesmax= -(1/beta)*np.log(histo+10*err)
                #fesmax -= offset
                #fesmin= -(1/beta)*np.log(histo-10*err)
                #fesmin -= offset
                feserr = (1/beta)*err/histo
                #axs[counter1,counter2].fill_between(bin_centers,fes-10*feserr,fes+10*feserr,color=colors[counter1],linewidth=0,alpha=0.5)
                popt, pcov = curve_fit(func, bin_centers[np.isfinite(fes)], fes[np.isfinite(fes)], sigma=2*feserr[np.isfinite(fes)])
                #x=np.linspace(histmin,histmax,10)
                #axs[counter1,counter2].plot(x,func(x,*popt),'--',color='black',alpha=0.5)
                axs[counter1].scatter(temp,popt[0],color=colors[counter1])
                chem_pot=np.insert(chem_pot,0,popt[0])
                chem_pot_temps=np.insert(chem_pot_temps,0,temp)
        counter2 += 1
        axs[counter1].plot([235,295],[0,0],'--',color='black',alpha=0.25,zorder=0)
        axs[counter1].set_xlim([235,295])
        axs[counter1].set_xticks([240,290])
        axs[counter1].set_ylim([-0.4,0.4])
    popt, pcov = curve_fit(func2, chem_pot_temps, chem_pot, sigma=np.ones(chem_pot.shape[0])*0.1, absolute_sigma=True)
    perr = np.sqrt(np.diag(pcov))
    x=np.linspace(temps[0]-10,temps[-1]+10,10)
    axs[counter1].plot(x,func2(x,*popt),'--',color='black',alpha=0.5)
    melt_entropy=np.insert(melt_entropy,0,popt[0])
    melt_entropy=np.insert(melt_entropy,0,popt[0])
    melt_temp=np.insert(melt_temp,0,popt[1])
    melt_temp_pressures=np.insert(melt_temp_pressures,0,press)
    melt_temp_errors=np.insert(melt_temp_errors,0,perr[1])
    axs[counter1].fill_between(x,func2(x,popt[0],popt[1]-perr[1]),func2(x,popt[0],popt[1]+perr[1]),color='black',alpha=0.1)
    counter1 += 1
    
print(melt_temp,melt_temp_pressures,melt_temp_errors)
#print(melt_temp_pressures,melt_entropy)

#for i in range(pressures.shape[0]):
#    axs[i,temps.shape[0]-1].text(210,2,str(pressures[i]/1000)+ ' kbar',rotation=90,va='center',ha='center',color=colors[i])
#for i in range(temps.shape[0]):
#    axs[0,i].text(160,15,str(temps[i])+ ' K',va='center',ha='center')

for i in range(pressures.shape[0]):
    axs[i].text(270,0.45,str(pressures[i]/1000)+ ' kbar',va='center',ha='center',color=colors[i])

#plt.subplots_adjust(wspace=0.45,hspace=0.2)
axs[2].set_xlabel("Temperature (K)")
axs[0].set_ylabel("Chemical potential (kJ/mol)")

plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
x=np.linspace(-2,2,1000)
```

```python
def fes(T,h,x):
    Tc=1
    r1=4
    s0=1
    G=-h*x+r1*((T-Tc)/Tc)*np.power(x,2) + s0*np.power(x,4)
    G -= np.amin(G)
    return G
```

```python
T=1.5
plt.plot(x,fes(T=T,h=1,x=x))
plt.plot(x,fes(T=T,h=0,x=x))
plt.plot(x,fes(T=T,h=-1,x=x))
plt.ylim([0,5])
plt.xlim([-2,2])

plt.show()
```

```python
T=1.1
plt.plot(x,fes(T=T,h=1,x=x))
plt.plot(x,fes(T=T,h=0,x=x))
plt.plot(x,fes(T=T,h=-1,x=x))
plt.ylim([0,5])
plt.xlim([-2,2])

plt.show()
```

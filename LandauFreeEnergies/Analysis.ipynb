# Model free energies from Landau theory
### For Figure 1

```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
x=np.linspace(-2,2,1000)
```

```python
def fes(T,h,x):
    Tc=1
    r1=4
    s0=1
    G=-h*x+r1*((T-Tc)/Tc)*np.power(x,2) + s0*np.power(x,4)
    G -= np.amin(G)
    return G
```

```python
font = {'family' : 'serif',
        'serif'   : 'palatino',
        #'sans-serif'    : 'Computer Modern Sans serif',
        'style'   : 'normal',
        'variant'   : 'normal',
        'stretch'   : 'normal',
        'weight'   : 'normal',
        'size'   : 20}
plt.rc('font', **font)
plt.rc('text', usetex=True)
plt.rcParams['figure.figsize'] = (3.0*3,1.5*3)
```

```python
mylinewidth=5
```

```python
T=1.5
plt.plot(x,fes(T=T,h=1,x=x),linewidth=mylinewidth)
plt.plot(x,fes(T=T,h=0,x=x),linewidth=mylinewidth)
plt.plot(x,fes(T=T,h=-1,x=x),linewidth=mylinewidth)
plt.ylim([-0.1,3])
plt.xlim([-2,2])

plt.savefig("very-sup.pdf")
plt.show()
```

```python
T=1.1
plt.plot(x,fes(T=T,h=1,x=x),linewidth=mylinewidth)
plt.plot(x,fes(T=T,h=0,x=x),linewidth=mylinewidth)
plt.plot(x,fes(T=T,h=-1,x=x),linewidth=mylinewidth)
plt.ylim([-0.1,3])
plt.xlim([-2,2])

plt.savefig("slightly-sup.pdf")

plt.show()
```

```python
T=0.5
plt.plot(x,fes(T=T,h=1,x=x),linewidth=mylinewidth)
plt.plot(x,fes(T=T,h=0,x=x),linewidth=mylinewidth)
plt.plot(x,fes(T=T,h=-1,x=x),linewidth=mylinewidth)
plt.ylim([-0.1,3])
plt.xlim([-2,2])

plt.savefig("sub.pdf")

plt.show()
```

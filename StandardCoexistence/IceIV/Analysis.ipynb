```python
import numpy as np
import matplotlib.pyplot as plt
#from joblib import Parallel, delayed
import time as tm
import itertools
#from progressbar import progressbar as pbar
from scipy.optimize import curve_fit
```

```python
font = {'family' : 'serif',
        'serif'   : 'palatino',
        #'sans-serif'    : 'Computer Modern Sans serif',
        'style'   : 'normal',
        'variant'   : 'normal',
        'stretch'   : 'normal',
        'weight'   : 'normal',
        'size'   : 35}
plt.rc('font', **font)
plt.rc('text', usetex=True)
plt.rcParams['figure.figsize'] = (18, 10)

colors = plt.rcParams['axes.prop_cycle'].by_key()['color']
```

```python
Tgrow = np.array([250,250,250,260,260,260,260,270,270])
Pgrow = np.array([2500,3000,3500,2500,3000,3500,4000,3500,4000])
Tmelt = np.array([270,270,280,280,280])
Pmelt = np.array([2000,2500,2500,3000,3500])
```

```python
Tcoexist=np.array([265,270,275])225.625
Pcoexist=np.array([2500,3000,3500])
```

```python
plt.scatter(Pgrow,Tgrow,marker='^',color=colors[0],label="Ice IV grows",s=200)
plt.scatter(Pmelt,Tmelt,marker='v',color=colors[0],label="Ice IV melts",s=200)
def func(x,a,b):
    return a*x+b

popt, pcov = curve_fit(func, Pcoexist, Tcoexist)
#plt.plot(Pcoexist,Tcoexist,color=colors[0],linewidth=2,label="L-Ice IV coexistence")
x=np.linspace(1500,4500,10)
plt.plot(x,func(x,*popt),color=colors[0],linewidth=4,label="L-Ice IV coexistence")
plt.fill_between([2600,3000],[240,240],[255,255],alpha=0.5,color=colors[1],label="L-L critical point")
plt.scatter([3200],[235],color=colors[1],label="HDL-LDL coexistence",s=200)
plt.plot([3500,3200,2800],[225.625,235,247.5],color=colors[1],label="HDL-LDL coexistence",linewidth=4)
plt.legend(bbox_to_anchor=(1.04,1), loc="upper left")
plt.text(3000,272,"Liquid-Ice IV coexistence",color=colors[0],rotation=12,va='center',ha='center')
plt.text(3120,241,"HDL-LDL coexistence",color=colors[1],rotation=-34,va='center',ha='center')
plt.xlabel("Pressure (bar)")
plt.ylabel("Temperature (K)")
plt.xlim([1900,4100])
plt.ylim([230,285])

plt.show()
```

```python
plt.scatter([3000,2500],[263,254],color=colors[0],label="L-Ice XIII coexistence",s=200)
def func(x,a,b):
    return a*x+b

popt, pcov = curve_fit(func, [3000,2500],[263,254])
#plt.plot(Pcoexist,Tcoexist,color=colors[0],linewidth=2,label="L-Ice IV coexistence")
x=np.linspace(1500,4500,10)
plt.plot(x,func(x,*popt),color=colors[0],linewidth=4,label="L-Ice XIII coexistence")

plt.fill_between([2600,3000],[240,240],[255,255],alpha=0.5,color=colors[1],label="L-L critical point")
plt.scatter([3200],[235],color=colors[1],label="HDL-LDL coexistence",s=200)
plt.plot([3500,3200,2800],[225.625,235,247.5],color=colors[1],label="HDL-LDL coexistence",linewidth=4)
plt.legend(bbox_to_anchor=(1.04,1), loc="upper left")
plt.text(3000,266,"Liquid-Ice XIII coexistence",color=colors[0],rotation=21,va='center',ha='center')
plt.text(3120,241,"HDL-LDL coexistence",color=colors[1],rotation=-34,va='center',ha='center')
plt.xlabel("Pressure (bar)")
plt.ylabel("Temperature (K)")
plt.xlim([1900,4100])
plt.ylim([230,285])

plt.show()
```

```python
import numpy as np
from scipy.optimize import curve_fit
import matplotlib.pyplot as plt
```

```python
font = {'family' : 'serif',
    'serif'   : 'palatino',
    'style'   : 'normal',
    'variant'   : 'normal',
    'stretch'   : 'normal',
    'weight'   : 'normal',
    'size'   : 30}
plt.rc('font', **font)
plt.rc('text', usetex=True)
plt.rcParams['figure.figsize'] = (4*3,3*3)
```

```python
def func(x,a,b):
    return a*x+b

data=np.genfromtxt("iceIV.txt")

popt, pcov = curve_fit(func, data[:,0], data[:,1])

print("Slope of ice IV with HDL is ", popt[0])

data=np.genfromtxt("iceV.txt")

popt, pcov = curve_fit(func, data[:,0], data[:,1])

print("Slope of ice V with HDL is ", popt[0])
```

```python
def func(x,a,b):
    return a*x+b

x = np.linspace(0.05,0.25,100)

data=np.genfromtxt("iceIV.txt")

plt.scatter(data[:,0], data[:,1],label="Ice IV")

popt, pcov = curve_fit(func, data[:,0], data[:,1])

plt.plot(x,func(x,*popt))

print("Slope of ice IV with HDL is ", popt[0], " +- ", np.sqrt(np.diag(pcov))[0])

data=np.genfromtxt("iceV.txt")

plt.scatter(data[:,0], data[:,1],label="Ice V")

popt, pcov = curve_fit(func, data[:,0], data[:,1])

plt.plot(x,func(x,*popt))

print("Slope of ice V with HDL is ", popt[0], " +- ", np.sqrt(np.diag(pcov))[0])

data=np.genfromtxt("kinks.txt")

plt.scatter(data[:,0], data[:,1],label="Kinks")

popt, pcov = curve_fit(func, data[:,0], data[:,1])

plt.plot(x,func(x,*popt))

print("Slope of LDL-HDL is ", popt[0], " +- ", np.sqrt(np.diag(pcov))[0])


plt.xlabel("Pressure (GPa)")
plt.ylabel("Temperature (K)")
plt.legend()

plt.show()
```
